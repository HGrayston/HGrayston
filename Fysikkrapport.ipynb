## Introduksjon

Det skal undersøkes hvordan en kule i en berg- og dalbane vil rulle.Rullebevegelsen til en kule på et krumt underlag blir annerledes enn hvis underlaget hadde vært rett. På et krumt underlag vil hastighet og akselerasjon alltid endre seg med tiden. Vi ønsker å finne den gjennomsnittlige slutthastigheten til kula. 

## Teori

Newtons 2. lov og energibevarelse kan brukes til å utlede en rekke størrelser som hastighet, akselerasjon og krefter for objekter. Total mekanisk energi hos et objekt er bevart dersom arbeidet på objektet er utført av konservative krefter, $E=K+U$. Den kinetiske energien tilsvarer da summen av translasjons- og rotasjonsenergi [2](https://www.ntnu.no/wiki/download/attachments/98076114/06-Potensiell%20energi.pdf?version=1&modificationDate=1481530811000&api=v2).

Objekter som ruller langs krum bane har et treghetsmoment tilsvarende $I_0=cmr^2$ hvor c=2/5, m er kulens masse, og r er kulens radius med hensyn på rotasjonsaksen gjennom massesenteret CM. [1](https://home.phys.ntnu.no/brukdef/undervisning/fyslab/files/labligninger.pdf) Rotasjonsenergi kan skrives som $\frac{1}{2}m(ωr)^2$ hvor $v=ωr$, $ω$ er vinkelhastighet til rotasjonen. Ligning for kinetisk energi er da gitt ved \eqref{eq:1}

\begin{align} \label{eq:1}
    K= \frac{1}{2}*m*v^2 + \frac{1}{2}*I_0*ω^2=\frac{7}{10}*m*v^2
    \tag{1}
\end{align}

Kulen starter i en høyde $y_0$ med null hastighet, total mekanisk energi er da lik kulens potensielle energi. $E = U_0 = mgy_0$ der g er tyngdeakselerasjonen. 

kulens høyde ved et gitt tidspunkt er y(x) og er en funksjon av kulens posisjon i x-retning. Et uttrykk for farten til kulen kan dermed gis som en funksjon av x som i ligning \eqref{eq:2}

\begin{align} \label{eq:2}
v(x) = \sqrt{\frac{2*g*(y_0-y(x))}{1+c}}
\tag{2}
\end{align}

Banens helningsvinkel er gitt ved $β = arctan(\frac{dy}{dx})$

Krumningen $k$ til banen, altså den inverse krumningsradien er gitt ved ligning \eqref{eq:3}
[1](https://home.phys.ntnu.no/brukdef/undervisning/fyslab/files/labligninger.pdf)

\begin{align} \label{eq:3}
κ=\frac{y''}{(1+(y')^2)^\frac{3}{2}}
\tag{3}
\end{align}

Sentripetalakselerasjonen er gitt ved $a_s =v^2κ$ [1](https://home.phys.ntnu.no/brukdef/undervisning/fyslab/files/labligninger.pdf), som med innsatte uttrykk for henholdsvis fart og krumning gir ligning \eqref{eq:4}

\begin{align} \label{eq:4}
a_s=\frac{2*g*(y_0-y)}{7}*\frac{y''}{(1+(y')^2)^\frac{3}{2}}
\tag{4}
\end{align}

Ved å se på Newtons 2. lov kan man finne et uttrykk for normalkraften N. 

\begin{align} \label{eq:5}
ma_s=N-m*g*cos(β)
\tag{5}
\\ \label{eq:6}
N = m*(g*cos(β) + a_s)
\tag{6}
\end{align}

Et krav for at kulen skal rulle rent er at friksjonskraften f er statisk. Altså må $μ_s>=|\frac{f}{N}|$, der $μ_s$ er det statiske friksjonstallet mellom banen og kulen [1](https://home.phys.ntnu.no/brukdef/undervisning/fyslab/files/labligninger.pdf). 

Som figuren over viser er det to krefter som virker tangentielt på kulen. Friksjonskraften f og tyngdens tangentialkomponent $-Mgsin(β)$. Newtons 2. lov gir da at $f-mgsin(β) = ma$ hvor $a=\frac{dv}{dt}$.
Spinnsatsen gir at $f=-\frac{I_0(\frac{dω}{dt})}{r}=-cma$.
Ved innsetting gir det ligning \eqref{eq:7} 

\begin{align} \label{eq:7}
f=\frac{2*m*g*sin(β)}{7}
\tag{7}
\end{align}

Horisontal hastighetskomponent er gitt ved $v_x = vcos(β)$. Farten kan og skrives som $v_x=\frac{Δx}{Δt}$. Gjennomsnittlig horisontal hastighetskomponent på et intervall n er gitt ved $v_{x,n}=\frac{1}{2}(v_{x,n-1} + v_{x,n})$.
Ved å kombinere dette kan et uttrykk for tidssteget utledes ved ligning \eqref{eq:8}

\begin{align} \label{eq:8}
Δt_n = \frac{Δx_n}{v_{x,n}} = \frac{2*Δx_n}{v_{x,n-1}+v_{x,n}}
\tag{8}
\end{align}

Verdien til t(x) kan da beregnes ved å summere verdiene av $Δt$ over hver $Δx$ frem til x-posisjonen. 


Ved analyse av data kan gjennomsnittet u av N antall målinger estimeres ved ligning \eqref{eq:9}

\begin{align} \label{eq:9}
u_{gjennomsnitt}=\frac{1}{N}\Sigma_{i=1}^{N}u_i
\tag{9}
\end{align}

Dette er et estimat for den mest sannsynlige verdien for størrelsen som er målt. Standardavviket er et mål for usikkerheten for hver måling, og kan beregnes ved ligning \eqref{eq:10}, [3](https://home.phys.ntnu.no/brukdef/undervisning/fyslab/files/usikkerhet.pdf)

\begin{align} \label{eq:10}
δu=\sqrt{\frac{1}{N-1}\Sigma_{i=1}^{N}(u_i-u_{gjennomsnitt})^2}
\tag{10}
\end{align}

Standardfeilen er et mål på usikkerheten, og er gitt ved ligning \eqref{eq:11}, [3](https://home.phys.ntnu.no/brukdef/undervisning/fyslab/files/usikkerhet.pdf).

\begin{align} \label{eq:11}
δu_{gjennomsnitt}=\frac{δu}{\sqrt{N}}
\tag{11}
\end{align}

## Metode
Først begynner vi med å importere nyttige bibliotek som brukes for å generere banen. Blant annet bruker vi pyplot-biblioteket i matplotlib til å tegne grafene for banefarten. I tillegg bruker vi en funksjon som heter CubicSpline fra interpolate-biblioteket til scipy. Denne bruker vi for å interpolere data ved hjelp av kubiske spliner, som er stykkevise 3. grads polynomer som til sammen beskriver en glatt linje. $(kilde kubiske spliner)$ Trenger kilde.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import CubicSpline
from scipy.signal import savgol_filter
```

Deretter lager vi en array xfast som inneholder x-koordinatene til festepunktene. Horisontal avstand mellom festepunktene er 0.200 m, og den betegner vi som $h$.

```python
h = 0.200
xfast=np.asarray([0,h,2*h,3*h,4*h,5*h,6*h,7*h])
```

Vi definerer variablene $xmin$, $xmax$ og $dx$ for henholdsvis start, slutt og steglengde i x-retning:

```python
xmin = 0.000
xmax = 1.401
dx = 0.001
```

Vi har en array $xfast$ for x-koordinatene, men vi trenger også en array med høydekoordinatene til festepunktene. Den kan vi definere som arrayen $yfast$. 

```python
xfast = [0., 0.2, 0.4, 0.6, 0.8, 1., 1.2, 1.4]
yfast = [0.273, 0.227, 0.200, 0.195, 0.156, 0.110, 0.126, 0.151]
print(np.array(xfast)*100)
print(np.array(yfast)*100 - 27.3 + 63) # Nullpunkt
```

Som beskrevet i $(kilde kubiske spliner)$ beregner programmet 7 tredjegradspolynomer, et for hvert intervall mellom to nabofestepunkter, med scipy.interpolate-funksjonen CubicSpline. Vi kaller denne funksjonen for $cs$, og definerer den på intervallet xfast, yfast:

```python
cs = CubicSpline(xfast, yfast, bc_type='natural')

xmin = 0.000
xmax = 1.401
dx = 0.001
```

Funksjonen cs kan nå brukes til å regne ut $y(x)$, $y'(x)$ og $y''(x)$ for en vilkårlig horisontal posisjon $x$, eventuelt for mange horisontale posisjoner lagret i en tabell: <br>
`cs(x)`   tilsvarer $y(x)$<br>
`cs(x,1)` tilsvarer $y'(x)$<br>
`cs(x,2)` tilsvarer $y''(x)$<br>


Vi vil ha en tettere diskretisering av x og y, så vi lager en tabell med x-verdier mellom $xmin = 0 \mathrm m$ og $xmax = 1.4 \mathrm m$.

```python
x = np.arange(xmin, xmax, dx) 
```

Funksjonen arange returnerer verdier på det "halvåpne" intervallet
`[xmin,xmax)`, dvs slik at xmin er med mens xmax ikke er med. Her blir
dermed `x[0]=xmin=0.000`, `x[1]=xmin+1*dx=0.001`, ..., `x[1400]=xmax-dx=1.400`, 
dvs x blir en tabell med 1401 elementer
<br>

<par> Vi lager arrays for $y$, $y'$ og $y''$ -- også med 1401 elementer. </par>

```python
Nx = len(x)
y = cs(x)       #y=tabell med 1401 verdier for y(x)
dy = cs(x,1)    #dy=tabell med 1401 verdier for y'(x)
d2y = cs(x,2)   #d2y=tabell med 1401 verdier for y''(x)
```

Nå kan vi plotte baneformen $y(x)$. Plottet for dette er vist i figur $(baneform)$ nedenfor

```python
baneform = plt.figure('y(x)',figsize=(12,6))
plt.plot(x,y,xfast,yfast,'*')
plt.title('Banens form')
plt.xlabel('$x$ (m)',fontsize=20)
plt.ylabel('$y(x)$ (m)',fontsize=20)
plt.grid()
plt.show()
```

Vi kan også plotte andre interessante størrelser, som for eksempel hellningsvinkelen i antall grader $\beta$ mot horisontal forflytning i antall meter $x$. Ved å bruke definisjonen vi fant for $\beta$ i seksjon $(teori)$, kan vi regne ut $\beta$ og plotte denne mot $x$. Da får man plottet som er vist i figur $(hellningsvinkel)$

```python
g = 9.81
M = 30.3/1000

beta = np.arctan(dy)
```

```python
def plot(x, y, xlabel, ylabel, title, options='', label=''):
    plt.figure(title,figsize=(8,5))
    assert len(x) == len(y)
    if isinstance(x[0], (list, tuple, np.ndarray)):
        for i in range(len(x)):
            alpha = 1
            if i != 0:
              alpha = 0.3
            if isinstance(options, (list, tuple, np.ndarray)):
                plt.plot(x[i], y[i], options[i], label=label[i], alpha=alpha)
            else:
                plt.plot(x[i], y[i], options, label=label[i], alpha=alpha)
        plt.legend(loc="upper left")
    else:
        plt.plot(x, y, options, label=label)
    plt.title(title)
    plt.xlabel(xlabel,fontsize=16)
    plt.ylabel(ylabel,fontsize=16)
    plt.grid()
    plt.show()
```

```python
a = (-5*g*np.sin(beta))/7


f = (2*M*g*np.sin(beta))/7
```

```python
plot(x, beta, 'x (m)', 'beta (grader)', 'Hellningsvinkel')
```

Ved å bruke krumningsformelen som er gitt i $(teori)$, kan vi også plotte krumning $K(x)$ målt i $\frac{1}{\mathrm m}$ mot horisontal forflytning $x$ målt i antall meter m.

Dermed får man et krumningsplott som vist i figur $(krumningsplott)$ nedenfor.

```python
k = d2y/(1+(dy)**2)**(3/2)


plot(x, k, 'x (m)', 'K(x) (1/m)', 'Krumningsplot')
```

Videre kan vi også plotte farten $v(x)$ målt i $\frac{\mathrm m}{\mathrm s}$ som en funksjon av horisontal posisjon x målt i antall meter m. Plottet for dette er vist i figur $(fart\ ved\ posisjon)$ nedenfor.

```python
v_x = np.sqrt((2*g*(y[0] - y))/(1 + 2/5))

plot(x, v_x, 'x (m)', 'v(x) (m/s)', 'Farten som en funksjon av horisontal posisjon x')
```

Det går også an å plotte kreftene som en funksjon av horisontal forflytning $x$. Her er normalkraften $N$ målt i newton plottet mot horisontal forflytning $x$ målt i antall meter m $(normalkraft)$,

```python
a_s=v_x**2*k
N=M*(g*np.cos(beta)+a_s)

plot(x, N, 'x (m)', 'N (N)', 'Normalkraft plot')
```

og her er friksjonskraften $f$ målt i $|\mathrm f/ \mathrm N|$ plottet mot horisontal forflytning $x$ målt i antall meter m $(friksjonskraft)$.

```python
f = np.abs((2*M*g*np.sin(beta))/7)

plot(x, f, 'x (m)', '|f/N|', 'Friksjonskraften som en funksjon av horisontal posisjon x')
```

Det går også an å plotte både posisjon og fart med hensyn på tid. I figur $(posisjon)$ er horisontal posisjon $x$ i antall meter m plottet mot tiden $t$ i sekunder.

```python
a=-((5*g)/7)*np.sin(beta)
```

```python
t_n = np.zeros(Nx)
    
for n in range(1, Nx):
    t_n[n] = t_n[n-1] + (2 * dx) / (v_x[n-1] + v_x[n])

plot(t_n, x, 't (s)', 'x (m)', 'Tidsutvikling av x')
```

I figur $(fart)$ er farten $v$ i $\frac{\mathrm m}{\mathrm s}$ plottet mot tiden $t$ i sekunder

```python
v = np.sqrt((10*g*(y[0]-y))/7)
t_n = np.zeros(Nx)
for i in range(1, Nx):
    v[i] = v_x[i]/np.cos(beta[i])
    t_n[i] = t_n[i-1] + (2 * dx) / (v_x[i-1] + v_x[i])

plot(t_n, v, 't (s)', 'v (m/s)', 'Tidsutvikling av fart')
```

Målt slutthastighet og endelig kinetisk energi er bestemt,

Etter at banen ble satt opp, ble det målt sluttfart på ballen 10 ganger ved hjelp av Tracker. Rådata ble hentet ut av programmet og ble plottet på graf. Koden under viser til fremgangsmåten for å omgjøre rådata til funksjoner i figur $(v_n(t))$. Hensikten var å kunne beregne målefeilen, og det kan allerede ses på grafen. Beregnet fart er betraktelig høyere f.o.m. $t=1.2s$, og lavere enn målingene fra $t=0$ t.o.m. $t=0.6s$. 

```python
def readFile(path):
    with open(f"drive/MyDrive/nytt/{path}", 'r') as f:
        data = np.genfromtxt(f, delimiter = ",",dtype=float,skip_header=2,usemask=True)
    return data[:, 0], data[:, 1], data[:, 2], data[:, 3] # t, x, y, v



x_values = [t_n[::16]]
v_values = []
v_values_filtered = [v[::16]]
labels = ['Beregnet fart']
for i in range(1, 11):
    t_measured,x_measured,y_measured,v_measured = readFile(f'data{i}.txt')

    idx = np.isfinite(t_measured) & np.isfinite(v_measured)
    v_values.append(v_measured[idx])
    #https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.savgol_filter.html
    # Lav-pass filter
    filterd_values = savgol_filter(v_measured[idx], 11, 3)
    x_values.append(t_measured[idx])
    v_values_filtered.append(filterd_values)
    labels.append(f'data{i}')

plot(x_values, v_values_filtered, 't (s)', 'v() (m/s)', 'Tidsutvikling av fart', options='', label=labels)


```

De ble gjort rede for avvik og feil i målingene som ble hentet ut ved å ta utgangspunkt i sluttfarten. Det ble også gjort rede for den kinetiske sluttenergien. Gjennomsnittsfarten for 10 målinger av $v_n(t) \frac{m}{s}$ ble målt til å være $1.262 \frac{m}{s}$. Det leseren kan bemerke seg av er at gjennomsnittlig sluttfart er betraktelig lavere enn den beregnede sluttfarten. Dette gir et fint frampek til utregning av standardavviket, som lå på $±0.0222 \frac{m}{s}$. Dette gir også opphav for standardfeilen som ble $±0.0070 \frac{m}{s}$. Det er ikke like innlysende å se det på figur, men ved bruk av formelen for kinetisk energi og den gjennomsnittlige sluttfarten, ble den gjennomsnittlige kinetiske sluttenergien regnet ut til å være lik $0.0242 J$.


```python
#Gjennomsnitt for N målinger av x (fart)

final_speed = [v[-1] for v in v_values]


average_final_speed = sum(final_speed)/(len(final_speed))
print("Gjennomsnittlig slutthastighet: ", '%.4f' % average_final_speed, "m/s")
print("Simulert slutthastighet: ", '%.4f' % v[-1], "m/s")
```

```python
#E=1/2*m*v**2

E_k_values = [0.5*M*v**2 for v in final_speed]

avg_E_k = 0.5*M*average_final_speed**2
print('%.4f' % avg_E_k, "J")
```

```python
#Standardavvik for s

def standardavvik(values, avg):
  sum_new = 0
  for v in values:
      sum_new+=(v-avg)**2

  s=np.sqrt((1/(len(values)-1))*sum_new)
  return s
s_v = standardavvik(final_speed, average_final_speed)
s_E_k = standardavvik(E_k_values, avg_E_k)
print('%.4f m/s' % s_v)
print('%.4f J' % s_E_k)

```

```python
#Standardfeil
def standardfeil(s, N):
  return s/np.sqrt(N)
s_feil_v = standardfeil(s_v, len(final_speed))
s_feil_E_k = standardfeil(s_E_k, len(E_k_values))
print('%.4f m/s' % s_feil_v)
print('%.4f J' % s_feil_E_k)
```

## Diskusjon

Hvis vi sammenligner simuleringen med dataene, ser vi at fartsutviklingen samsvarer godt med forsøkene. Mot slutten kan man samtid se at den simulerte farten er noe høyere. Dette kan skyldes noe høyere friksjon i banen, som f.eks. at ballen triller sidelengs, eller at banen ikke er montert helt etter målene som simuleres. Dette stemmer også med den gjennomsnittlige sluttfarten og den simulerte sluttfarten. 

En annen feilkilde ligger i programmet Tracker som ble brukt til å hente ut data om ballens forflytning. I Tracker bestemmer man et koordinatsystem og hvor lang én meter er i pixler. Dette gir grunnlaget for alle utregninger av x-verdier, y-verdier og fartsverdier i datasettet. Kameraets optikk, i hovedsak brennvidden til kamera, vil dermed være viktig for at dataene skal bli rett. Større brennvidde vil resultere i at bildet blir forvrengt og den fysisk avstanden representeres ved færre pixler i kantene på bilde. For å minimere brennvidens påvirkning ble kameraet plassert lengst mulig unna, uten at det gikk ut over kvaliteten. Målingene, som er i form av videoer, er begrenset til 60 bilder i sekundet. Dette gjør at Tracker ikke kan regne ut farten i tiden mellom bildene. Dette kan resultere i feil utregning, og gjør at dataene er taggete.

Standardavviket forteller oss hvor stor spredning det er i datasettet. Standardavviket for de ti forsøkene som ble gjennomført er $\pm0.0222\frac{m}{s}$. Det er altså lav spredning mellom forsøkene og forventingsverdien.
